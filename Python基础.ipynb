```python
x = [ 1,2,3,4,8,7, 22,33, 88]
print("原数据：",x)
for i in range(len(x)):
    if(x[i] % 2) !=0:   # 判读第 i 个元素是否为奇数
        x[i] = x[i] * x[i]
print("处理后：",x)
```

##### 【例2-40】使用argsort和lexsort函数进行排序。

```python
#对名称进行排序：首先按姓氏，然后按名称。
surnames = ('Hertz', 'Galilei', 'Hertz')
first_names = ('Heinrich', 'Galileo', 'Gustav')

ind = np.lexsort((first_names, surnames)) # Sort by surnames, then by first_names，原数组不变

print(ind)

#array([1, 2, 0])
print([surnames[i] + ", " + first_names[i] for i in ind])
```

```python
#对两列数字进行排序：
a = [1,5,1,4,3,4,4] # First column

b = [9,4,0,4,0,2,1] # Second column

ind = np.lexsort((b,a)) # Sort by a, then by b,原数组不变

ind

#array([2, 0, 4, 6, 5, 3, 1])
print([(a[i],b[i]) for i in ind])
print(a)
print(b)
```

```python
arr = np.array([7,9,5,2,9,4,3,1,4,3])
print('原数组：',arr)
print('排序后：',arr.argsort())  #返回值为数组排序后的下标排列
print(arr) #依旧不会改变原数组
a=[2,5,8,4,3,7,6]
b=[9,4,0,4,0,2,1]
c=np.lexsort((a,b)) 
print(c)
```

    inplace=False：查询是应该修改数据还是返回修改后的副本
    kwargs：dict关键字参数

3.  返回值类型：

    DataFrame，由提供的查询表达式产生的DataFrame

```python
df = pd.DataFrame({'key1' : ['a', 'a', 'b', 'b', 'a'],
    'key2' : ['yes', 'no', 'yes', 'yes', 'no'],
    'data1' : np.random.randn(5),
    'data2' : np.random.randn(5)})
grouped = df['data1'].groupby(df['key1']) #按照key1分组 df['data1']数据分组后返回的数据类型是一个groupby对象
print(df)
print(grouped.size()) #调用对象groupby的方法 size(),返回一个含有分组大小的series
print(grouped.mean())  #调用对象groupby的方法 mean(),返回每个分组数据的均值
print(grouped.max())  #调用对象groupby的方法 size(),返回每个分组数据的最大值
```

##### 【例2-77】通过DataFrame最后一列的数值进行正负分组。

按函数分组：函数作为分组键的原理类似于字典，通过映射关系进行分组，但是函数更加灵活。

箱线图（Boxplot）也称箱须图，其绘制需使用常用的统计量，能提供有关数据位置和分散情况的关键信息，尤其在比较不同特征时，更可表现其分散程度差异。箱线图利用数据中的五个统计量（最小值、下四分位数、中位数、上四分位数和最大值）来描述数据，它也可以粗略地看出数据是否具有对称性、分布的分散程度等信息，特别可以用于对几个样本的比较。 

**boxplot函数**格式：

matplotlib.pyplot.boxplot(x, notch=None, sym=None, vert=None, whis=None, positions=None, widths=None, patch_artist=None,meanline=None, labels=None, … )


##### 【例2-95】boxplot绘图示例。

```python
import pandas as pd
np.random.seed(2)  #设置随机种子
#df = pd.DataFrame(np.random.rand(1,500)) #生成0-1之间的1*500维度数据并存入 DataFrame中
df = pd.DataFrame(np.random.rand(5,4), columns = ['A', 'B', 'C', 'D']) #生成0-1之间的5*4维度数据并存入4列DataFrame中
plt.boxplot(df)
plt.show()
```

#### 概率图

概率图模型是图灵奖获得者Pearl提出的用来表示变量间概率依赖关系的理论。  
**正态分布又名高斯分布。**

正态概率密度函数normpdf（x，mu，sigma），其中，x为向量，mu为均值，sigma为标准差。


```python
from scipy.stats import norm
fig,ax = plt.subplots()
plt.rcParams['font.family'] = ['SimHei'] 
np.random.seed(1587554)
mu = 100  #设置均值
sigma = 15  #设置标准差
x = mu+sigma*np.random.randn(437) 
num_bins = 50
n,bins,patches = ax.hist(x,num_bins,density = 1)
y=norm.pdf(bins,mu,sigma)
ax.plot(bins,y,'--')
fig.tight_layout()
plt.show()
```
